# Serie de Fourier
- Se presentan los conceptos bsicos de la serie de Fourier.

- Sea $x(t)\in \mathbb{R},\mathbb{C}$ una seal definida en el intervalo finito $t\in[t_i,t_f]$ o $t\in[-T/2,T/2]$, que cumple las [condiciones de Dirichlet](https://en.wikipedia.org/wiki/Dirichlet_conditions#:~:text=In%20mathematics%2C%20the%20Dirichlet%20conditions,point%20where%20f%20is%20continuous.):

  - $\int_T |x(t)| dt < \infty$.
  - $x(t)$ presenta discontinuidades hasta de primer orden.
  - $x(t)$ presenta una cantidad finita de mnimos y mximos.

- La representacin generalizada de Fourier de $x(t)$ se puede definir a partir de un conjunto de funciones base ortogonales $\{\phi_n(t)\in\mathbb{R},\mathbb{C}\}_{n\in\mathbb{Z}}$, tal que:

$$
\int_T \phi_n(t)\phi_m^*(t)dt =\left\{\begin{align}
0; & \quad n\neq m\\ E_n=\int_T|\phi_n(t)|^2dt; & \quad n=m\end{align}\right.,
$$

mediante la combinacin lineal:

$${\hat{x}}(t)=\sum_{n=-N}^N c_n \phi_n(t)$$

- A partir de la minimizacin de la potencia media del error $\bar{P}_e = \frac{1}{T}\int_T |x(t)-\hat{x}(t)|^2dt,$ los pesos $c_n\in \mathbb{R,C}$, denominados el espectro, se pueden calcular como:


$$c_n = \frac{\langle x(t),\phi_n(t) \rangle}{\|\phi_n(t)\|_2^2} = \frac{\int_T x(t)\phi^*_n(t)dt}{\int_T |\phi_n(t)|^2dt}.$$


- Adems, la potencia media del error, $\bar{P}_e$, se puede reescribir como:

$$\bar{P}_e = \bar{P}_x-\sum^N_{n=-N}|c_n|^2P_n \geq 0$$

- Entonces:

$$\bar{P}_x \geq \sum^N_{n=-N}|c_n|^2P_n$$

- La anterior desigualdad se conoce como la [desigualdad de Bessel](https://en.wikipedia.org/wiki/Bessel%27s_inequality).

- Para el caso en el que $N\to \infty,$ el [teorema de Parseval](https://en.wikipedia.org/wiki/Parseval%27s_theorem) establece que:


$$\lim_{N\to \infty} \bar{P}_x-\sum^N_{n=-N}|c_n|^2P_n = 0$$.

- En este caso:

$$P_n = \frac{1}{T}\int_T |\phi_n(t)|^2dt$$

**Nota**: A partir de la potencia media del error, se puede definir una medida normalizada, denominada error relativo, como:

$$e_r[\%] = \frac{\bar{P}_e}{\bar{P_x}} 100[\%] = \left(1-\frac{\sum^N_{n=-N}|c_n|^2P_n}{\bar{P}_e}\right)100[\%]$$





# Serie de Fourier exponencial

- Para la serie de Fourier exponencial, las bases se fijan como $\{\phi_n(t)=e^{jn\omega_o t}\}_{n\in\mathbb{Z}}$, por ende la reconstruccin de la seal se escribe segn:

$$\hat{x}(t)=\sum_{n=-N}^N c_ne^{jn\omega_o t}$$

donde $\omega_o = \frac{2\pi}{tf-ti}=\frac{2\pi}{T}.$

- Adems, teniendo en cuenta que:

$$
\int_T e^{jn\omega_ot}e^{-jm\omega_ot}dt =\left\{\begin{align}
0; & \quad n\neq m\\ E_n=\int_T|e^{jn\omega_o t}|^2dt=T; & \quad n=m\end{align}\right.,
$$

- El espectro $c_n$ se puede calcular como:

$$c_n = \frac{1}{T}\int_T x(t)e^{-jn\omega_ot}dt$$

- El espectro para $n=0$, se conoce comunmente como el nivel DC u offset (promedio) de la seal $x(t)$, ya que:

$$c_0 = \frac{1}{T}\int_T x(t)e^{0}dt=\frac{1}{T}\int_T x(t)dt.$$

**Nota:** Dado que $E_n=T$ para las bases de la serie de Fourier $e^{jn\omega_ot},$ la potencia $P_n = 1.$

- El espectro $c_n$, nos da una idea del parecido entre la seal de entrada $x(t)$ y la base $e^{jn\omega_o t}$, dado el grado de variacin segn la frecuencia de las bases  $\omega_n = n\omega_o,$ donde $n$ se conoce como el *armnico* o mltiplo de la frecuencia fundamental $\omega_o$.

- Por lo general, es de inters obtener el grfico de magnitud y fase de $c_n$, donde:

$$|c_n|=\sqrt{{\rm{Re}}^2\{c_n\}+{\rm{Im}}^2\{c_n\}}$$

y

$$\theta_n = \tan^{-1}\left(\frac{{\rm{Im}}\{c_n\}}{{\rm{Re}}\{c_n\}}\right)$$.

- En general, se habla del contenido frecuencial de la seal segn el espectro de frecuencia $c_n$, y es ampliamente utilizado en ingeniera (por ejemplo ver [Diagramas de Bode](https://en.wikipedia.org/wiki/Bode_plot)).

# Serie de Fourier trigonomtrica y compacta

- Por facilidad de clculo, la serie exponencial de Fourier se puede extender a sus versiones trigonomtrica y compacta.

- La serie trigonomtrica asume que:

$$\hat{x}(t) = \sum_{n=0}^{N}a_n\cos(n\omega_o t) + b_n\sin(n\omega_o t)$$.

- Aplicando el mismo razomiento de representacin mediante bases ortogonales, se tiene que:

$$a_n = \frac{\langle x(t),\cos(n\omega_ot)\rangle}{\|\cos(n\omega_ot)\|_2^2}=\frac{\int_Tx(t)\cos(n\omega_ot)dt}{\int_T|\cos(n\omega_ot)|^2dt}$$

$$a_n = \frac{2}{T}\int_Tx(t)\cos(n\omega_ot)dt$$

y

$$b_n = \frac{\langle x(t),\sin(n\omega_ot)\rangle}{\|\sin(n\omega_ot)\|_2^2}=\frac{\int_Tx(t)\sin(n\omega_ot)dt}{\int_T|\sin(n\omega_ot)|^2dt}$$

$$b_n = \frac{2}{T}\int_Tx(t)\sin(n\omega_ot)dt$$

- Para el caso particular $n=0$:

$$a_0 = \frac{\int_Tx(t)\cos(0)dt}{\int_T|\cos(0)|^2dt}=\frac{\int_Tx(t)dt}{\int_Tdt}$$

$$a_0 =c_0 = \frac{1}{T}\int_Tx(t)dt$$

$$b_0 = \frac{\int_Tx(t)\sin(0)dt}{\int_T|\sin(0)|^2dt}\to 0$$

$$b_0 = 0$$

- Dado que $x(t)$ puede presentar simetra par o impar, $x(t)=x(-t)\to$ SIMETRA PAR, $x(t)=-x(-t)\to$ SIMETRA IMPAR, y dado que el $\cos(\theta)$ presenta simetra PAR y el $\sin(\theta)$ presenta simetra IMPAR, la serie trigonomtrica puede simplificar sus clculos para cuando $x(t)$ presenta algn tipo de simetra (Ver [material apoyo](https://github.com/amalvarezme/SenalesSistemas/blob/master/3_SerieyTransformadaFourier/Simetria_SerieFourier.pdf)).

- Finalmente, la serie trigonomrica compacta, resume la representacin utilizando funciones coseno desfasadas:

\begin{equation}
d_n = \sqrt{a_n^2+b_n^2} = 2|c_n|; \quad n > 0
\end{equation}
\begin{equation}
d_0 = a_0;  
\end{equation}

\begin{equation}
\phi_n = \tan^{-1}\left(\frac{\mathsf{Im}\{c_n\}}{\mathsf{Re}\{c_n\}}\right) = \tan^{-1}\left(\frac{-b_n}{a_n}\right); \quad n \geq 0
\end{equation}

y:

\begin{equation}
\hat{x}(t) = \sum\limits_{n=0}^N{d_n \cos(n\omega_ot+\phi_n)}.
\end{equation}

## Implementaciones

- Si $x(t)$ presenta algun tipo de simetra, se sugiere utilizar la serie trigonomtrica de Fourier, y para pasar a la exponencial, se puede utilizar la siguiente relacin:

\begin{equation}
c_n = \frac{a_n - jb_n}{2}, \quad n\neq 0\\
c_0 = a_0.
\end{equation}

- Si se desea calcular la serie de Fourier exponencial, para pasar a su versin trigonomtrica se puede utilizar la siguiente relacin:

$$a_0 = c_0, \quad b_0 = 0$$

$$a_n = 2{\rm{Re\{c_n\}}}$$

$$b_n = -2{\rm{Im\{c_n\}}}$$


## Ejemplo

- Sea la seal:

\begin{equation}
x(t) = \begin{cases} A, & -T/2 \leq t < 0 \\ -A, & 0 \leq t < T/2 \end{cases}
\end{equation}

- Su espectro en serie de Fourier (compleja), se calcula como:

\begin{equation}
c_n = \frac{1}{T}\int^{T/2}_{-T/2}{x(t)e^{-jn\omega_o t}}dt=\frac{jA}{n\pi}(1-(-1)^n); \quad n\neq 0
\end{equation}

donde

\begin{equation}
c_0 = \frac{1}{T}\int^{T/2}_{-T/2}{x(t)}dt=0
\end{equation}
y

\begin{equation}
\hat{x}(t) = \sum\limits_{n=-N}^N{c_n e^{jn\omega_ot}}.
\end{equation}

Adems, su espectro en serie de Fourier (trigonomtrica) se calcula como:

\begin{equation}
a_n = \frac{2}{T}\int^{T/2}_{-T/2}{x(t)\cos(n\omega_o t)}dt=2\mathsf{Re}\{c_n\}= 0; \quad n > 0
\end{equation}
\begin{equation}
a_0 = c_0 = 0.
\end{equation}

\begin{equation}
b_n = \frac{2}{T}\int^{T/2}_{-T/2}{x(t)\sin(n\omega_o t)}dt=-2\mathsf{Im}\{c_n\}=\frac{2A}{n\pi}((-1)^n-1); \quad n > 0
\end{equation}
\begin{equation}
b_0 = 0.
\end{equation}

y:

\begin{equation}
\hat{x}(t) = a_0+\sum\limits_{n=1}^N{a_n \cos(n\omega_ot)+b_n\sin{n\omega_ot}}.
\end{equation}


Por otra parte, su espectro en serie de Fourier compacta se calcula como:

\begin{equation}
d_n = \sqrt{a_n^2+b_n^2} = 2|c_n| = \frac{2A}{n\pi}((-1)^n-1); \quad n > 0
\end{equation}
\begin{equation}
d_0 = a_0;  
\end{equation}

\begin{equation}
\phi_n = \tan^{-1}\left(\frac{\mathsf{Im}\{c_n\}}{\mathsf{Re}\{c_n\}}\right) = \tan^{-1}\left(\frac{-b_n}{a_n}\right); \quad n \geq 0
\end{equation}

- A continuacin, se presentan las simulaciones respectivas.

```python
# Para graficar
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")

mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

A = 1 # amplitud de la seal
T = 5 # segundos de grfica
Fs = 100 # frecuencia de muestreo definida
tv = np.arange(-T/2,T/2,1/Fs) #vector de tiempo generado
Nm = len(tv) #numero de muestras requeridas

#se crea la seal A y -A concatenando
xo = np.concatenate((A*np.ones(int(Nm/2)), -A*np.ones(int(Nm/2))), axis=0)
#graficar seal original
plt.plot(tv,xo,'r',linewidth=4)
plt.grid()
plt.xlabel("t[s]",fontsize = 14)
plt.ylabel("x(t)",fontsize = 14)
plt.show()
```

- Para ilustrar las bases de Fourier, se crea una matriz con la cantidad de armnicos deseados segn la variable $N$. Cada base se guarda como una columna de la matriz `phiN`.


```python
N = 10 #numero armonicos
wo = 2*np.pi/T # frecuencia fundamental
# definir bases
phin = np.zeros((Nm,2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervalo de inters
```

```python
phin.shape #tamao matriz phin
```

```python
# graficar bases
from ipywidgets import interact,IntSlider
fig, ax = plt.subplots(1, 1,figsize=(6,3))
def pltbase(n=1):

    plt.plot(tv,np.real(phin[:,int(n+N)]),label="Re{$\phi_n$(t)}")
    plt.xlabel("t[s]",fontsize = 14)
    plt.ylabel("$\phi_n$(t)",fontsize = 14)
    plt.plot(tv,np.imag(phin[:,int(n+N)]),label="Im{$\phi_n$(t)}")
    plt.grid()
    plt.plot(tv,xo,'r',linewidth=4,label='$x(t)$')
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.gcf().set_figwidth(6)
    plt.gcf().set_figheight(3)
    plt.show()
    return
interact(pltbase,n=(-N,N,1))
```

```python
from matplotlib import animation, rc
from IPython.display import HTML

#figura y axes animacinx
fig, ax = plt.subplots(1, 1,figsize=(6,3))
#funcin para grfico interactivo
def update(n):
    #actualizacin regresor Bayesiano
    #Graficar prior
    ax.clear()
    ax.plot(tv,np.real(phin[:,int(n+N)]),label="Re{$\phi_n$(t)}")
    ax.set_xlabel("t[s]",fontsize = 12)
    ax.set_ylabel("$\phi_n$(t)",fontsize = 12)
    ax.plot(tv,np.imag(phin[:,int(n+N)]),label="Im{$\phi_n$(t)}")
    ax.plot(tv,xo,'r',linewidth=4,label='$x(t)$')
    ax.set_title(f'Bases armnico n = {n}')
    ax.legend(loc='lower left')
    plt.show()

#desplegar animacin
anim = animation.FuncAnimation(fig, update, frames=np.arange(-N,N,1), blit=False)
rc('animation', html='jshtml')
```

# Definicin y grfica del espectro de la serie de Fourier con numpy

La siguiente celda de cdigo codifica el espectro $c_n$ segn el ejemplo resuelto.



```python
#calcular espectro -> cambiar segn la seal estudiada
cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N, N, 2*N+1) #vector num armnicos
Cn = ((1j*A)/(nv*np.pi))*(1-(-1)**nv)
Cn[N] = 0 # nivel dc C_0
```

```python
#graficar espectro
fig = plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.stem(nv, np.real(Cn), 'r')
plt.xlabel(r'$n$',fontsize = 14)
plt.ylabel(r'$Re\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,2)
plt.stem(nv, np.imag(Cn), 'r')
plt.xlabel(r'$n$',fontsize = 14)
plt.ylabel(r'$Im\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,3)
plt.stem(nv, abs(Cn), 'r')#magnitud del espectro
plt.xlabel(r'$n$',fontsize = 14)
plt.ylabel(r'$|C_n|$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(2,2,4)
plt.stem(nv, np.angle(Cn), 'r')#fase del espectro
plt.xlabel(r'$n$',fontsize = 14)
plt.ylabel(r'$\langle C_n$',fontsize = 14)
plt.axis('tight')
plt.grid()

fig.tight_layout()
```

- Para calcular el error de reconstruccin respecto a la cantidad de armnicos considerados (filtrado espectral), se clcula como:

$$e_r[\%] = \frac{\bar{P}_e}{\bar{P_x}} 100[\%] = \left(1-\frac{\sum^N_{n=-N}|c_n|^2P_n}{\bar{P}_e}\right)100[\%]$$

- En este caso, la potencia de $x(t)$ se clcula como:

\begin{equation}
P_x = \int\limits_{-T/2}^{T/2}{|x(t)^2|dt} = A^2
\end{equation}
para el caso de la serie exponencial de Fourier $\varphi_n(t)=e^{jn\omega_ot}$ y $P_n=1$, por consiguiente:

\begin{equation}
E_r[\%] = 1 - \left(\frac{1}{P_x}\sum\limits_{n=-N}^N{|c_n|^2}\right).
\end{equation}

- A continuacin, se clcula el error relativo en funcin de los armnicos considerados y se muestra la seal filtrada para cada caso.


```python
Px = A**2 # estimar segn seal estudiada
def pltest(Na=1):
    ind = np.arange(N-Na,N+Na+1)
    er = 1- np.sum(abs(Cn[ind])**2)/Px
    #seal reconstruida
    xe = phin[:,ind].dot(Cn[ind] ) # dot calcula el producto entre vectores
    plt.plot(tv,xe,color='b',label='$x_e(t)$') # seal estimada o filtrada
    plt.plot(tv,xo,color='r',label='x(t)') #seal original
    plt.title('$E_r$=%.2f[%%]' % (100*er),fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()

    return
interact(pltest,Na=(1,N,1))
```

```python
#figura y axes animacinx
fig, ax = plt.subplots(2, 3,figsize=(12,5))
#funcin para grfico interactivo
def update(n):
    #actualizacin regresor Bayesiano
    #Graficar prior
    ax[0,0].clear()
    ax[0,1].clear()
    ax[0,2].clear()
    ax[1,0].clear()
    ax[1,1].clear()
    ax[1,2].clear()

    #espectro
    ind = np.arange(N-n,N+n+1) #armnicos a considerar
    ax[0,0].stem(nv, abs(Cn), 'b')#magnitud del espectro
    ax[0,0].stem(nv[ind], abs(Cn[ind]), 'r')#magnitud del espectro
    ax[0,0].set_xlabel(r'$n$[rad/s]')
    ax[0,0].set_ylabel(r'$|C_n|$')

    #componentes espectrales
    ce = phin[:,ind].dot(np.diag(Cn[ind]) )
    ax[0,1].plot(tv,np.real(ce))#componente real
    ax[0,1].set_xlabel('t[s]')
    ax[0,1].set_ylabel('$Re\{c_n\phi_n(t)\}$')

    ax[0,2].plot(tv,np.imag(ce))#componente imaginario
    ax[0,2].set_xlabel('t[s]')
    ax[0,2].set_ylabel('$Im\{c_n\phi_n(t)\}$')

    #reconstruccion
    er = 1- np.sum(abs(Cn[ind])**2)/Px
    #seal reconstruida
    xe = phin[:,ind].dot(Cn[ind] ) # dot calcula el producto entre vectores
    #error en t
    et = xo - xe

    ax[1,1].plot(tv,xo,'r',linewidth=4,label='$x(t)$')
    ax[1,1].plot(tv,xe,color='b',label='$x_e(t)$') # seal estimada o filtrada
    ax[1,1].plot(tv,et,color='g',label='$e(t)$') # seal estimada o filtrada
    ax[1,1].set_title('$E_r$=%.2f[%%]' % (100*er))
    ax[1,1].set_xlabel('t[s]')
    ax[1,1].set_ylabel('x(t)')
    ax[1,1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax[1,0].axis('off')
    ax[1,2].axis('off')
    plt.subplots_adjust(wspace=0.5,hspace=0.7)
    plt.show()

#desplegar animacin
anim = animation.FuncAnimation(fig, update, frames=np.arange(0,N+1,1), blit=False)
rc('animation', html='jshtml')
```

- En este sentido, si se evalua la reconstruccin por fuera del intervalo finito definido $[t_i,t_f]$, la serie de Fourier replica la reconstruccin (asume seal peridica $x(t)$).

1. Presente una simulacin relacionada con la proyeccin lineal de $\mathbf{a}\in\mathbb{R}^2$ sobre $\mathbf{b}\in\mathbb{R}^2$. El cdigo debe permitir al usuario ingresar las componentes de los dos vectores, calcular el factor de escala de la proyeccin y graficar los vectores de entrada y el proyectado (usar la funcin [pyplot.arrow](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.arrow.html)).

2. Encontrar la representacin por serie exponencial de Fourier de la seal $x(t) = A {\rm{rect}}_{\tau/2}(t)$, con $t\in [-\tau/2,\tau/2]$. Graficar el espectro, la fase del espectro, la reconstruccin para $n\in\{0,\pm1,\pm2,\dots\}$, y determinar el valor del error relativo para cada valor de $n$.
```python
#Punto 1. Simulacin relacionada con la proyeccin lineal de a sobre b

import numpy as np
import matplotlib.pyplot as plt #Llamamos a la libreria

def proyec(a, b): #Funcin para calcula la proyeccin de a sobre b
    b_norm = np.linalg.norm(b)
    if b_norm == 0:
        raise ValueError("El vector b no debe ser el vector cero")
    escal = np.dot(a, b) / np.dot(b, b)
    proy = escal * b
    return proy, escal

def graf_vec(a, b, proy): #Funcin para graficar los vectores
    plt.figure()
    plt.axvline(0, color='black', lw=0.5)
    plt.axhline(0, color='black', lw=0.5)


    plt.arrow(0, 0, a[0], a[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue', label='Vector a') #vector a
    plt.arrow(0, 0, b[0], b[1], head_width=0.1, head_length=0.1, fc='red', ec='red', label='Vector b') #vector b
    #Vector proyeccin
    plt.arrow(0, 0, proy[0], proy[1], head_width=0.1, head_length=0.1, fc='green', ec='green', label='Proyeccin de a sobre b')

    plt.grid()
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Ingreso de los usuarios
a = np.array([float(x) for x in input("Ingrese las componentes del vector a (separadas por un espacio): ").split()])
b = np.array([float(x) for x in input("Ingrese las componentes del vector b (separadas por un espacio): ").split()])

proy, escal = proyec(a, b) #calculo de la proyeccin y factor escala

#Resultados
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Proyeccin de a sobre b: {proy}")
print(f"Factor de escala de la proyeccin: {escal}")

graf_vec(a, b, proy) #Graficar los vectores
```

```python
#Ejercicio punto 2. Con la ayuda

#Graficar el espectro, la fase del espectro, la reconstruccin para  n{0,1,2,}
#determinar el valor del error relativo para cada valor de n.
#Utilice el ambiente animacin de Python para mostrar los resultados.

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

# Parmetros de la seal
A = 2  # Amplitud de la seal
T = 8  # Duracin en segundos
Fo = 1 / T  # Frecuencia fundamental
Fs = 100 * Fo  # Frecuencia de muestreo
tv = np.arange(-T / 2, T / 2, 1 / Fs)  # Vector de tiempo generado
Nm = len(tv)  # Nmero de muestras

# Generacin de la seal rectangular
xo = np.zeros(Nm)
xo[(tv >= -T / 4) & (tv <= T / 4)] = A

# Graficar seal original
plt.figure(figsize=(10, 4))
plt.plot(tv, xo, 'r', linewidth=4)
plt.grid()
plt.xlabel("t[s]", fontsize=14)
plt.ylabel("x(t)", fontsize=14)
plt.title("Seal Rectangular")
plt.show()

# Nmero de armnicos
N = 10
wo = 2 * np.pi / T  # Frecuencia fundamental

# Definir bases de Fourier
phin = np.zeros((Nm, 2 * N + 1), dtype=np.complex_)
for n in range(-N, N + 1):
    phin[:, n + N] = np.exp(1j * n * wo * tv)  # Bases de Fourier

# Animacin de las bases armnicas
fig, ax = plt.subplots(1, 1, figsize=(6, 3))
def update(n):
    ax.clear()
    ax.plot(tv, np.real(phin[:, int(n + N)]), label="Re{$\phi_n$(t)}")
    ax.plot(tv, np.imag(phin[:, int(n + N)]), label="Im{$\phi_n$(t)}")
    ax.plot(tv, xo, 'r', linewidth=4, label='$x(t)$')
    ax.set_xlabel("t[s]", fontsize=12)
    ax.set_ylabel("$\phi_n$(t)", fontsize=12)
    ax.set_title(f'Bases armnicas n = {n}')
    ax.legend(loc='lower left')
    plt.grid()

anim = animation.FuncAnimation(fig, update, frames=np.arange(-N, N + 1), blit=False)
rc('animation', html='jshtml')
HTML(anim.to_html5_video())

# Clculo del espectro de Fourier
cn = np.zeros(2 * N + 1, dtype=np.complex_)
nv = np.linspace(-N, N, 2 * N + 1)  # Vector nmero armnicos
Cn = A * np.sin(0.5 * nv * np.pi) / (nv * np.pi)
Cn[N] = 0.5 * A  # Nivel DC C_0

# Graficar espectro
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].stem(nv, np.real(Cn), 'r')
axs[0, 0].set_xlabel(r'$n$', fontsize=14)
axs[0, 0].set_ylabel(r'$Re\{C_n\}$', fontsize=14)
axs[0, 0].grid()

axs[0, 1].stem(nv, np.imag(Cn), 'r')
axs[0, 1].set_xlabel(r'$n$', fontsize=14)
axs[0, 1].set_ylabel(r'$Im\{C_n\}$', fontsize=14)
axs[0, 1].grid()

axs[1, 0].stem(nv, abs(Cn), 'r')
axs[1, 0].set_xlabel(r'$n$', fontsize=14)
axs[1, 0].set_ylabel(r'$|C_n|$', fontsize=14)
axs[1, 0].grid()

axs[1, 1].stem(nv, np.angle(Cn), 'r')
axs[1, 1].set_xlabel(r'$n$', fontsize=14)
axs[1, 1].set_ylabel(r'$\langle C_n$', fontsize=14)
axs[1, 1].grid()

fig.tight_layout()
plt.show()

# Reconstruccin de la seal y clculo del error
Px = 0.5 * A**2  # Estimar segn seal estudiada

def pltest(Na=1):
    ind = np.arange(N - Na, N + Na + 1)
    er = 1 - np.sum(abs(Cn[ind])**2) / Px
    xe = phin[:, ind].dot(Cn[ind])  # Seal reconstruida
    plt.figure(figsize=(10, 4))
    plt.plot(tv, xe, color='b', label='$x_e(t)$')  # Seal estimada
    plt.plot(tv, xo, color='r', label='x(t)')  # Seal original
    plt.title(f'$E_r$=%.2f[%%]' % (100 * er), fontsize=16)
    plt.xlabel('t[s]')
    plt.ylabel('x(t)')
    plt.grid()
    plt.legend()
    plt.show()

from ipywidgets import interact
interact(pltest, Na=(1, N, 1))

# Animacin de la reconstruccin
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
def update(n):
    ax[0, 0].clear()
    ax[0, 1].clear()
    ax[0, 2].clear()
    ax[1, 0].clear()
    ax[1, 1].clear()
    ax[1, 2].clear()

    ind = np.arange(N - n, N + n + 1)  # Armnicos a considerar
    ax[0, 0].stem(nv, abs(Cn), 'b')  # Magnitud del espectro
    ax[0, 0].stem(nv[ind], abs(Cn[ind]), 'r')  # Magnitud del espectro
    ax[0, 0].set_xlabel(r'$n$')
    ax[0, 0].set_ylabel(r'$|C_n|$')

    ce = phin[:, ind].dot(np.diag(Cn[ind]))
    ax[0, 1].plot(tv, np.real(ce))  # Componente real
    ax[0, 1].set_xlabel('t[s]')
    ax[0, 1].set_ylabel('$Re\{c_n\phi_n(t)\}$')

    ax[0, 2].plot(tv, np.imag(ce))  # Componente imaginario
    ax[0, 2].set_xlabel('t[s]')
    ax[0, 2].set_ylabel('$Im\{c_n\phi_n(t)\}$')

    er = 1 - np.sum(abs(Cn[ind])**2) / Px
    xe = phin[:, ind].dot(Cn[ind])  # Seal reconstruida
    et = xo - xe

    ax[1, 1].plot(tv, xo, 'r', linewidth=4, label='$x(t)$')
    ax[1, 1].plot(tv, xe, color='b', label='$x_e(t)$')  # Seal estimada
    ax[1, 1].plot(tv, et, color='g', label='$e(t)$')  # Error
    ax[1, 1].set_title(f'$E_r$=%.2f[%%]' % (100 * er))
    ax[1, 1].set_xlabel('t[s]')
    ax[1, 1].set_ylabel('x(t)')
    ax[1, 1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
    ax[1, 0].axis('off')
    ax[1, 2].axis('off')
    plt.subplots_adjust(wspace=0.5, hspace=0.7)
    plt.show()

anim = animation.FuncAnimation(fig, update, frames=np.arange(0, N + 1, 1), blit=False)
rc('animation', html='jshtml')
HTML(anim.to_html5_video())
```
